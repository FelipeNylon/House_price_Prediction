```python
#imports 
import pandas as pd 
from sklearn import linear_model as lm 
import numpy as np
```

```python
#loading data
data = pd.read_csv(r"C:\Users\Lipe Marlon\Documents\House_Rockets\kc_house_data.csv")

data[0:9]
```

```python
#data preparation 
x_train = data.drop(['price', 'date'], axis=1)
y_train = data['price'].copy()
display(x_train.head())
```

```python
#model training 

model_lr = lm.LinearRegression()

model_lr.fit(x_train,y_train)
```

```python
pred = model_lr.predict(x_train)
pred[0:100]
```

```python
#perfomance metrics
datacopy = data.copy()
```

```python
datacopy['prediction'] = pred
```

```python
datacopy[['id', 'price', 'prediction', 'error', 'error_abs','error_percentual','error_percentual_abs']][0:9]
```

```python
#calculando o erro da regressÃ£o linear 

datacopy['error'] = datacopy['price'] - datacopy['prediction']
datacopy['error_abs'] = np.abs(datacopy['error'])
datacopy['error_percentual'] = ((datacopy['price']-datacopy['prediction'])/datacopy['price'])
datacopy['error_percentual_abs'] = np.abs(datacopy['error_percentual'])
```

```python
np.sum(datacopy['error_abs']) / len(datacopy['error_abs'])
```

```python
#tirando a mÃ©dia de error das prediÃ§Ãµes, 25% de mÃ©dia de erro da regressÃ£o 
np.mean(datacopy['error_percentual_abs'])
```
